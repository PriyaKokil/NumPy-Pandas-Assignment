```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = sns.load_dataset('planets')
```

```python
df.head()
```

## 1.Scatter plot: Visualize the relationship between 'orbital_period' and 'mass' of the planets.

```python
sns.scatterplot(x='orbital_period',y='mass',data=df)
plt.xlabel('orbital_period')
plt.ylabel('mass')
plt.title('orbital_period vs mass')
plt.show()
```

## 2.Bar plot: Display the count of planets discovered by each method.

```python
df.info()
```

```python
sns.countplot(x='method' , data = df)
plt.xticks(rotation=90)
plt.title("Count of Planets Discovered by Method")
plt.xlabel("Discovery Method")
plt.ylabel("Count")
plt.show()
```

## 3.Histogram: Visualize the distribution of 'distance' of planets from their respective stars.

```python
sns.histplot(x='distance' , data = df,bins=20,kde=True)
plt.title("Distribution of Distance of Planets from Stars")
plt.xlabel("Distance (parsecs)")
plt.ylabel("Count")
plt.show()
```

## 4. Pair plot: Show pairwise relationships between 'orbital_period', 'mass', 'year', and 'distance'.

```python
sns.pairplot(data = df[['orbital_period', 'mass', 'year', 'distance']])
plt.subtitle('pairwise relationships between orbital_period, mass, year and distance', y=1.02)
plt.show()
```

## 5.Violin plot: Compare the distribution of 'year' for different 'method' of planet discovery.

```python
sns.violinplot(x='method',y='year',data = df)
plt.title('distribution of year for different method of planet discovery')
plt.xticks(rotation=90)
plt.xlabel('method')
plt.ylabel('year')
plt.show()
```

## 6.Swarm plot: Visualize the 'mass' of planets discovered by different 'method'

```python
sns.swarmplot(x='method',y='mass',data=df)
plt.xticks(rotation=90)
plt.title("Mass of Planets Discovered by Different Methods")
plt.xlabel("Discovery Method")
plt.ylabel("Mass (Jupiter Mass)")
plt.show()
```
